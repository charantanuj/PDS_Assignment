```python

import pandas as pd
from scipy.stats import ttest_ind

# Reading the CSV file
data = "/content/cleaned_data_fratility.csv"
data_set = pd.read_csv(data)

print(data_set.describe())
```

```python
from scipy.stats import ttest_ind

data_set['Frailty'] = pd.Categorical(data_set['Frailty'])
frail = data_set[data_set['Frailty'] == 1]['Weight in lb']
non_frail = data_set[data_set['Frailty'] == 0]['Weight in lb']

t_statistic, p_value = ttest_ind(frail,non_frail, equal_var=False)


print(" Results:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Save to test File
result_file_path = "/content/result.txt"
with open(result_file_path, 'w') as file:
    file.write("T-statistic: {}\n".format(t_statistic))
    file.write("P-value: {}\n".format(p_value))
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x='Frailty', y='Grip_strength in kg', data=data_set)
plt.title('Grip Strength using Frailty')
plt.show()
```
